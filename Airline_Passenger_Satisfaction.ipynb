### A Chi-square test
 I perform a chi-square test to check relationship between categorical variables. The chi-square test examine if there is a real difference between groups.

H0 - There is no significant difference in satisfaction between female and male passengers.

HA -  There is a significant difference in satisfaction between female and male passengers.

The p-value is compared with a significance level of 0.05. 

If p < 0.05, the null hypothesis is rejected.

If p ≥ 0.05, the null hypothesis is not rejected.

```python
contingency_table = pd.crosstab(data['Gender'], data['satisfaction'])
contingency_table
```

```python
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(contingency_table)

chi2, p
```

`chi-square` = 1.65, A small value of chi-square means that the difference is not large.

`p-valu`e = 0.199, 0.199 > 0.05

H0 is not rejected. The difference between female and male passengers is small, and the chi-square test shows it is not significant.
